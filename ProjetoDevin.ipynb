```python
#Preparando o ambiente SPark
#Instalando biblioteca pyspark
!pip install pyspark


#Importando Bibliotecas
from pyspark.sql import SparkSession

#Criando objeto spark:
spark = SparkSession.builder.appName("SENAI").getOrCreate()

from pyspark.sql import functions as f
```

```python
#Importando data frame de um arquivo .csv:
url = "/content/drive/MyDrive/Devin/archive"

developers = spark.read.options(header=True).csv(
    url,
    inferSchema=True,
    sep=","
)
```

```python
developers = developers.withColumn("MeanSDSal_Adjus", developers.MeanSDSal_Adjus.cast("float"))\
                      .withColumn("MeanSDSal_Unajust", developers.MeanSDSal_Unajust.cast("float"))\
                      .withColumn("Mean_Ununadj_Sal_all_occup", developers.Mean_Ununadj_Sal_all_occup.cast("float"))\
                      .withColumn("Num_of_Sof_Dev_Jobs", developers.Num_of_Sof_Dev_Jobs.cast("float"))\
                      .withColumn("Median_Home_Price", developers.Median_Home_Price.cast("float"))\
                      .withColumn("Cost_of_Living_avg", developers.Cost_of_Living_avg.cast("float"))\
                      .withColumn("Rent_avg", developers.Rent_avg.cast("float"))\
                      .withColumn("Cost_of_LivPlus_Rent_avg", developers.Cost_of_LivPlus_Rent_avg.cast("float"))\
                      .withColumn("Local_Purch_Pow_avg", developers.Local_Purch_Pow_avg.cast("float"))
```

```python
developers.printSchema()
```

```python
developers.createOrReplaceTempView("v_developers")
```

```python
spark.sql(
"""
  select d.metro, sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust)
  from v_developers d
  GROUP BY metro,MeanSDSal_Adjus, MeanSDSal_Unajust
"""
).show()
```

```python
spark.sql(
"""
  select d.metro, sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust)
  from v_developers d
  GROUP BY metro, MeanSDSal_Adjus, MeanSDSal_Unajust
  order by d.metro
"""
).show()
```

```python
spark.sql(
"""
  select d.metro, sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust)
  from v_developers d
  GROUP BY metro, MeanSDSal_Adjus, MeanSDSal_Unajust
  order by d.metro
"""
).toPandas()
```

```python
spark.sql(
"""
  select count(d.metro), sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust), max(d.Num_of_Sof_Dev_Jobs)
  from v_developers d
  GROUP BY metro, MeanSDSal_Adjus, MeanSDSal_Unajust
  order by d.metro
"""
).toPandas()
```

```python
spark.sql(
"""
  select count(d.metro), sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust), max(d.Num_of_Sof_Dev_Jobs)
  from v_developers d
  GROUP BY metro, MeanSDSal_Adjus, MeanSDSal_Unajust, d.Num_of_Sof_Dev_Jobs
  order by Num_of_Sof_Dev_Jobs
"""
).toPandas()
```

```python
spark.sql(
"""
  select d.metro, sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust), max(d.Num_of_Sof_Dev_Jobs)
  from v_developers d
  GROUP BY metro, MeanSDSal_Adjus, MeanSDSal_Unajust, d.Num_of_Sof_Dev_Jobs
  order by Num_of_Sof_Dev_Jobs desc
"""
).toPandas()
```

```python
spark.sql(
"""
  select d.metro, sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust), max(d.Num_of_Sof_Dev_Jobs)
  from v_developers d
  GROUP BY metro, MeanSDSal_Adjus, MeanSDSal_Unajust, d.Num_of_Sof_Dev_Jobs
  order by Num_of_Sof_Dev_Jobs desc
"""
).toPandas()
```

```python
spark.sql(
"""
  select d.metro, d.city, sum(d.MeanSDSal_Adjus), avg(d.MeanSDSal_Unajust), max(d.Num_of_Sof_Dev_Jobs)
  from v_developers d
  GROUP BY metro, city, MeanSDSal_Adjus, MeanSDSal_Unajust, d.Num_of_Sof_Dev_Jobs
  order by Num_of_Sof_Dev_Jobs desc
"""
).toPandas()
```
