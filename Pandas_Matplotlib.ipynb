```python
dados = [
    ("iago", "assistente", 2900),
    ("ana", "asistente", 2780)
]
```

```python
# Unindo dois Data Frames:
# axis=0 --> Alterando linhas
# axis=1 --> Alterando colunas
df = pd.concat([df, df2], axis=0)
```

```python
# Deletando coluna:
df.drop("index", axis=1, inplace=True)
```

```python
# Criando cópia do Data frame:
copia = df.copy(deep=True)
```

```python
copia
```

```python
# Crie uma nova coluna chamada salario_reajustado contendo os reajustes conforme abaixo:
# Salário menor que 3500, reajuste de 7%
# Salário maior ou igual 3500, reajuste de 3%

salario_reajustado = []
```

```python
for i in df["salario"]:
  if i < 3500:
    salario_reajustado.append(i * 1.07)
  else:
    salario_reajustado.append(i * 1.03)
```

```python
salario_reajustado
```

```python
# Adicionando nova coluna no Pandas:
df["salario_reajustado"] = salario_reajustado
```

```python
# Importando arquivo csv:
url = "/content/drive/MyDrive/Devin/Pandas/moradias_SP.csv"

df = pd.read_csv(url)
```

```python
df
```

```python
# Dados estatísticos básicos:
df.describe()
```

```python
# Explorando dataframe
df.info()
```

```python
# Selecionando algumas colunas:
df[["district", "address"]]
```

```python
# Realizando filtros:
# 1° opção:
# Trazendo somente apartamentos com área construída acima de 100 mts

df[(df["type"] == 'Apartamento') & (df["area"] > 100)]
```

```python
# 2° forma --> Usando Series:
c1 = df["type"] == 'Apartamento'
c2 = df["area"] > 100
```

```python
c1
```

```python
# Operadores bitwise:
# and --> &
# or --> |
# not --> !

df[c1 & c2]
```

```python
# 3° forma de realizar filtros e pesquisas:
df.query("type == 'Apartamento' and area > 100")
```

```python
# Renomeando colunas do dataframe pandas:

df.rename(columns={
    "rent":"aluguel",
    "district":"bairro",
    "address":"endereco",
    "type":"tipo"
}, inplace=True)
```

```python
df
```

```python
# Normalizando dados strings:
df.fillna("N/A", inplace=True)

# Normalizando dados numéricos:
df.fillna(0, inplace=True)
```

```python
# Realizando agrupamentos e estatísticas:

# Mostre os 10 primeiros bairros que possuem a média de valor total maior:

df.groupby("bairro")[["total"]].mean().sort_values("total", ascending=False).head(10)
```

```python
# Mostrar quantas moradias disponíveis por tipo imóvel:

df.groupby("tipo")[["total"]].count().sort_values("total",ascending=False)
```

```python
# Gerando gráficos em pandas
# Uso da função plot()
df.plot()
```

```python
grafico = df.groupby("bairro")[["total"]].mean().sort_values("total", ascending=False).head()
```

```python
grafico.plot(
    kind="barh",
    color="darkblue",
    title="Valor médio dos 5 bairros mais caros de SP",
    xlabel="Valor médio",
    ylabel="Bairro",
    figsize=(5, 3)
)
```

```python
# Grando um gráfico de pizza para mostrar proporção por tipo de moradia:

grafico = df.groupby("tipo")[["total"]].count().sort_values("total",ascending=False)
```

```python
grafico.plot(
    kind="pie",
    title="Qtd de moradias por tipo",
    subplots=True,
    figsize=(5, 3),
    autopct="%1.1f%%",
    legend=None,
    colormap="summer"
)
```

```python
# Gerando gráfico matplotlib

import matplotlib.pyplot as plt2
from matplotlib import style
plt2.style.use("dark_background")
```

```python
# Gerando gráfico de dispersão --> Verificar se duas variáveis possuem correlação.
figura = plt2.figure(figsize=(5, 3))
eixo = figura.add_axes([0, 0, 1, 1])
eixo.set_title("Relação M2 vs Valor total")
eixo.set_xlabel("M2")
eixo.set_ylabel("Valor total")
eixo.scatter(df["area"], df["total"], color="white", marker=".")
```
